Preprocessing of modal split data (copied from PDF) into table format.

Source: Mobilität in Städten--SrV 2018, Tab 15 (a)

https://tu-dresden.de/bu/verkehr/ivs/srv/ressourcen/dateien/SrV2018_Staedtevergleich.pdf

1. First, copy table 15 (a) from the above pdf into a text file.
2. Use the below code to try and get it into reasonable shape.

```python
import re
```

```python
with open('../../data/raw/verkehrsanteile.txt', 'r') as f:
    txt = f.read()
```

```python
city = re.compile('[^\na-zA-Z]\s?([A-Z]+[^\d]*)')
city.findall(txt)
txt = city.sub(r'\n\1',txt)
```

```python
txt = re.sub('\s?%','',txt)
txt = re.sub('2019','',txt)
print(txt)
```

```python
separator = re.compile('\s(?=\d)')
txt = separator.sub('\t',txt)
print(txt)
```

```python
decimal = re.compile('(?<=\d),(?=\d)')
txt = decimal.sub('.',txt)
print(txt)
```

```python
headers = "Untersuchungsraum\tZu Fuß absolut\tFahrrad absolut\tMIV absolut\tÖV absolut\tGesamt\tZu Fuß relativ\tFahrrad relativ\tMIV relativ\tÖV relativ"
txt = headers + txt
print(txt)
```

```python
with open('../../data/interim/verkehrsanteile.csv', 'w') as f:
    f.write(re.sub('\t',',',txt))
```
