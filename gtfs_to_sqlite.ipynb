```python
import pandas as pd
from sqlalchemy import create_engine
import datetime as dt
import os
import re
import zipfile
```

```python
my_path = "../raw/gtfs/2021/"
my_file = my_path + '2021_reissue_2.zip'
my_id = "2021_210720"
```

```python
with zipfile.ZipFile(my_file,"r") as zip_ref:
    zip_ref.extractall(my_path)
```

```python
db_path = my_path + my_id + '.db'
if os.path.isfile(db_path):
    os.remove(db_path)

disk_engine = create_engine('sqlite:///'+ db_path)
```

```python
gtfs_re = re.compile("(^[a-z]+.*)\.txt")
# gtfs_files = {gtfs_re.match(f.name).groups()[0]: f.name for f in os.scandir(my_path) if gtfs_re.match(f.name)}
gtfs_files = [(gtfs_re.match(f.name).groups()[0], f.name) for f in os.scandir(my_path) if gtfs_re.match(f.name)]
```

```python
for k, v in gtfs_files:
    print(k)
```

```python
for name, fname in gtfs_files:
    print(fname)
    
    start = dt.datetime.now()
    chunksize = 200000
    j = 0
#     index_start = 1
    
    for df in pd.read_csv(my_path + fname, chunksize=chunksize, iterator=True, encoding='utf-8'):

#         df.index += index_start

        j+=1
        print(j)
        if j%10==0:
            print('\t{} seconds: completed {} rows'.format((dt.datetime.now() - start).seconds, j*chunksize))

        df.to_sql(name, disk_engine, if_exists='append')
#         index_start = df.index[-1] + 1
```
